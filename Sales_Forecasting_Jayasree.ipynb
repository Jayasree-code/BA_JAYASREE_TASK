# Importing Libraries

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.special import boxcox1p
import seaborn as sns
```

# Data Preparation
1. Get the data from .csv file.
2. Merging the data.
3. Analyze the data.
4. Manipulating the data.

```python
#Getting data
features=pd.read_csv("/content/drive/MyDrive/Sales_Forecasting_project/features.csv")
store=pd.read_csv("/content/drive/MyDrive/Sales_Forecasting_project/stores.csv")
train=pd.read_csv("/content/drive/MyDrive/Sales_Forecasting_project/train.csv")
test=pd.read_csv("/content/drive/MyDrive/Sales_Forecasting_project/test.csv")
```

#  Take important information and merging the data

From train.csv, taking groupby Store, Date and sum Weekly_Sales.

```python
#reseting train
train=train.groupby(['Store','Date'])['Weekly_Sales'].sum()
train=train.reset_index()
train.head()
```

# Merging train and features data by inner join.

```python
#merging train and feature
data=pd.merge(train,features,on=['Store','Date'],how='inner')
data.head(10)
```

# Merging Data and Store on Store by inner join.

```python
#merging store with data
data=pd.merge(data,store,on=['Store'],how='inner')
data.head(10)
```

# sorting the data by date.

```python
#sorting values of Data
data=data.sort_values(by='Date')
data.head(10)
```

# Box plot of Type and Weekly_Sales

# Manipulating data. Transform data into useful information and deleting unnecessary items. Getting the final data.

Removing the unnecessary columns

```python
data=data.drop(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'],axis=1)
data.head(10)
```

# Again checking for null values and verify that data is clean.

```python
data.isnull().sum()
```

# Heatmap for verifying there are no null values in data.

```python
sns.heatmap(data.isnull(),yticklabels=False, cmap="viridis")
```

# Converting IsHoliday in Holiday which is integer and 1 for holiday and 0 otherwise.

```python
data['Holiday']=[int(i) for i in list(data.IsHoliday)]
data.head(10)
```

# Converting type to an integer by one-hot encoding. The first column is also removed because we know of both columns B and C are 0 then it is A-type. So B=1 and C=0 for B.B=0 and C=1 for C.B=0 and C=0 for A.

```python
Type_dummy=pd.get_dummies(data['Type'],drop_first=True)
Type_dummy.head(10)
```

# Concating type_dummy with data.

```python
data=pd.concat([data,Type_dummy],axis=1)
data.head(10)
```

# We only use integer so we removing non-integer columns after transforming it into integer form

```python
data=data.drop(['Type','IsHoliday'],axis=1)
data.drop(10)
```

#  Now, we perform learning tasks on this data in four steps.
Splitting the train and test data.
Applying linear regression.
Predicting the value
Evaluate the model

# 1. Splitting data into train and test data. The size of the test data is 30%.

```python
#splitting data in input and output
X=data.drop(['Weekly_Sales','Store','Date'],axis=1)
y=data['Weekly_Sales']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

# 2. Applying linear regression and fit the training data into it.

```python
LR=LinearRegression(normalize=True)
LR.fit(X_train,y_train)
```

# 3. Predicting the data for test value as per linear regression.

```python
y_pred=LR.predict(X_test)
plt.plot(y_test,y_pred,'ro')
plt.plot(y_test,y_test,'b-')
plt.show()
```

# 4. Evaluating the model by calculating errors by the root mean square error and R -squared.

```python
Root_mean_square_error=np.sqrt(np.mean(np.square(y_test-y_pred)))
print(Root_mean_square_error)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(r2)
```

# Now, to predict the weekly sales, give particular tuple to input in the model and predict the weekly sales as output. 

```python
prediction=LR.predict(pd.DataFrame([(40.37,2.876,173.325456,7.934,103464,0,0,0)]))
print(prediction)
```
